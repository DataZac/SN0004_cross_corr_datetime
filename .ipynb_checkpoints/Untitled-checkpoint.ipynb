# MWE for using cross correlation to measure time delay between timeseries
# Some datetime object and time crunching snippets

```python
import numpy as np
import datetime as dt
from operator import itemgetter
import matplotlib.pyplot as plt
import math
import pandas as pd
import pprint
from scipy import signal
```

```python
time = np.arange(0*math.pi, 12*math.pi, 0.1)
def get_sin(x, upper = 4*math.pi, lower = 2*math.pi):
    if (x > lower) and (x < upper):
        return np.sin(x)
    else: 
        return 0

y = [get_sin(x) for x in time]
z = [get_sin(x, lower = 0*math.pi, upper = 2*math.pi) for x in time]
cro_cor_1 = signal.correlate(y,z, mode = 'full')
cro_cor_2 = signal.correlate(y,z, mode = 'valid')
cro_cor_3 = signal.correlate(y,z, mode = 'same')
print(np.argmax(y)-np.argmax(z))
print(len(time),len(cro_cor_1),len(cro_cor_2),len(cro_cor_3))

plt.scatter(x = [i for i in range(len(time))], y = y)
plt.show()
plt.scatter(x = [i for i in range(len(time))], y = z)
plt.show()
plt.scatter(x = [i-round(len(time)/2) for i in range(len(time))], y = cro_cor_3)
plt.show()
plt.scatter(x = [i for i in range(len(cro_cor_1))], y = cro_cor_1)
plt.show()
```

# MWE for aligning two time series of different freq.

```python
start = dt.datetime(2011, 1, 1, 23, 0, 0)
x = [[start+dt.timedelta(seconds=i),i+33+0.0,'speed', i] for i in range(4)]
freq = 10 # Hz
y = [[start+dt.timedelta(seconds=i/freq)+dt.timedelta(seconds=-0.13),i+44+0.1,'accel',i] for i in range(4*freq)]
y_cut = [ y[i]+[int(i/10)]  for i in range(len(y)) if ((i % 10) == 0) ]
pprint.pprint(y_cut)
#pd.date_range(start, end, period = 100)
pprint.pprint(x)
pprint.pprint(y)
```

```python
x = x+y
x = sorted(x, key =itemgetter(0))
print(x[0][0].timestamp())
pprint.pprint(x)    
```

```python
L = []
for i,datapoint in enumerate(x):
    if datapoint[2] == 'speed':
        if (datapoint[0]-x[i-1][0]) < (x[i+1][0]-datapoint[0]):
            L.append(x[i-1])
pprint.pprint(L)            
```
